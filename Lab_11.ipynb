# Travelling Salesman Problem using Dynamic Programming.
Problem Description:
Given a set of cities and distances, find the shortest possible route that visits each city once and returns to the starting city.

#Theory
TSP uses dynamic programming with bit masking. Store results of visited subsets to reduce repeated work.

## Algorithm
1. Start from source city
2. Mark visited cities using bit mask
3. Try all unvisited cities
4. Choose minimum cost path
5. Return to source city

```python
def tsp(graph):
    n = len(graph)
    dp = [[-1]*(1<<n) for _ in range(n)]

    def visit(city, mask):
        if mask == (1<<n) - 1:
            return graph[city][0]

        if dp[city][mask] != -1:
            return dp[city][mask]

        ans = float('inf')

        for next_city in range(n):
            if mask & (1<<next_city) == 0:
                new_cost = graph[city][next_city] + visit(next_city, mask | (1<<next_city))
                if new_cost < ans:
                    ans = new_cost

        dp[city][mask] = ans
        return ans

    return visit(0, 1)
```

```python
graph = [
[0, 10, 15, 20],
[10, 0, 35, 25],
[15, 35, 0, 30],
[20, 25, 30, 0]
]

print("Minimum travelling cost:", tsp(graph))
```

# Exercise
1. Print path of cities visited
2. Take graph input from user
3. Test with different number of cities

```python
def tsp_with_path(graph):
    n = len(graph)

    if n <= 1:
        return 0, [0] if n == 1 else []

    dp = [[-1] * (1 << n) for _ in range(n)]
    next_city_memo = [[-1] * (1 << n) for _ in range(n)]

    def visit(city, mask):
        if mask == (1 << n) - 1:
            return graph[city][0]

        if dp[city][mask] != -1:
            return dp[city][mask]

        ans = float('inf')
        best_next_city = -1

        for next_city in range(n):
            if mask & (1 << next_city) == 0:
                new_cost = graph[city][next_city] + visit(next_city, mask | (1 << next_city))

                if new_cost < ans:
                    ans = new_cost
                    best_next_city = next_city

        dp[city][mask] = ans
        next_city_memo[city][mask] = best_next_city
        return ans

    min_cost = visit(0, 1)

    path = [0]
    curr_city = 0
    curr_mask = 1

    for _ in range(n - 1):
        next_city = next_city_memo[curr_city][curr_mask]
        path.append(next_city)
        curr_mask |= (1 << next_city)
        curr_city = next_city

    path.append(0)

    return min_cost, path

def main():
    print("--- Travelling Salesman Problem Solver ---")

    try:
        n = int(input("Enter the number of cities: "))

        if n <= 1:
            print("Please enter a number of cities greater than 1.")
            return

        print(f"\nEnter the {n}x{n} distance matrix row by row (space-separated):")
        print("Example for row 1: 0 10 15 20")

        graph = []
        for i in range(n):
            row = list(map(int, input(f"Row {i+1}: ").split()))
            if len(row) != n:
                print(f"Error: Expected {n} values, but got {len(row)}.")
                return
            graph.append(row)

        cost, path = tsp_with_path(graph)

        print("\n--- Results ---")
        print(f"Minimum travelling cost: {cost}")

        path_str = " -> ".join(map(str, path))
        print(f"Path of cities visited: {path_str}")

    except ValueError:
        print("Invalid input. Please ensure you enter integers.")

if __name__ == "__main__":
    main()
```
